## Glossário

| Termo | Definição |
| :--- | :--- |
| Fraude | Transação financeira ilegítima, não autorizada pelo portador do cartão. |
| Transação Legítima | Transação financeira autorizada e realizada pelo portador do cartão. | 
| Desbalanceamento de Classes | Situação em que uma classe (fraude) tem muito menos exemplos que a outra (legítima). |
| Recall | Métrica que mede a proporção de fraudes reais que foram detectadas. |
| Precision | Métrica que mede a proporção das previsões de fraude que eram realmente fraudes. |
| Falso Positivo (FP) | Transação legítima erroneamente classificada como fraude. |
| Falso Negativo (FN) | Transação fraudulenta erroneamente classificada como legítima. |
| PCA (Análise de Componentes Principais) | Técnica de redução de dimensionalidade usada para anonimizar os dados. |
| SMOTE | Técnica de over-sampling que gera exemplos sintéticos da classe minoritária. |
| Over-sampling | Técnica que aumenta o número de exemplos da classe minoritária. |
| Under-sampling | Técnica que diminui o número de exemplos da classe majoritária. |
| Ajuste de Peso | Técnica que atribui pesos diferentes às classes durante o treinamento do modelo. |
| EDA (Análise Exploratória de Dados) | Processo de investigação inicial dos dados para descobrir padrões, anomalias e testar hipóteses. |
| Feature | Variável preditora usada pelo modelo de machine learning. |
| Modelo de Machine Learning | Algoritmo treinado para fazer previsões com base em dados históricos. |
| Trade-off | Situação em que se deve abrir mão de um aspecto para obter outro (ex: recall vs precision). |
